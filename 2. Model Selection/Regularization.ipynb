```python
import numpy as np
import matplotlib.pyplot as plt
```

Делаем датасет

```python
x = np.linspace(0, 2, 20)
y_no_noise =  x**3 + 2*x**2 + 4 
y = y_no_noise + np.random.normal(scale = 2, size = x.shape)
```

```python
plt.scatter(x, y)
plt.scatter(x, y_no_noise)
```

```python
def add_powers(x, max_power):
    return np.hstack([(x**i).reshape(-1, 1) for i in range(0, max_power)])
```

```python
data = add_powers(x, 20)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression(normalize=True)
```

```python
lr.fit(data, y)
```

```python
x_plot = np.linspace(0, 2, 100)
data_plot = add_powers(x_plot, 20)
y_plot = lr.predict(data_plot)
```

```python
plt.ylim(0,20)
plt.plot(x_plot, y_plot)
plt.scatter(x, y)
```
